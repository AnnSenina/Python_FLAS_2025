```python
print(ord('A')) # порядковый номер в Unicode от 0
print(chr(65))
```

ord() - функция, которая принимает строку Unicode из одного символа и возвращает значение кодовой точки

В Python кодовые точки Unicode могут быть записаны с помощью \u escape-последовательности, за которой следуют четыре шестнадцатеричные цифры, дающие кодовую точку

Еscape-последовательности - это и есть те самые символы из \ и буквы (или сочетания цифр), например '\n', '\t' и др.

```python
# '一' 1 на китайском (?)
print(ord('一')) # порядковый номер в Unicode
print(f'\\u{ord("一"):04x}') # python escape-последовательность Unicode
# \\ - экранируем \
# :04x в f-строке - перевод в шестнадцатеричную систему
```

```python
# китайские фразы из цифр
text = '''1314 yī sān yī sì
一生一世 yīshēng yīshì
На всю жизнь.

360 sān liù líng
想念你 xiǎngniàn nǐ
Скучаю по тебе.

25184 èr wǔ yī bā sì
爱我一辈子 ài wǒ yībèizi
Люби меня всю жизнь.'''

# давайте достанем число 1

import re
re.findall(r'一', text)
```

```python
re.findall(r'\u4e00', text) # python Unicode
```

```python
num = '''零
〇
一
二
三
四
五
六
七
八
九
十
廿
卅
卌
百
皕
千'''
# китайские числительные из википедии
```

```python
print(num.split())
```

```python
for i in num.split():
  print(ord(i), f'\\u{ord(i):04x}')

# есть ощущение, что находятся они вообще не по порядку
# \d или какой-то диапазон их не возьмет
```

```python
# но если вам нужно как-то извлечь их из текста, можно как-то так:
text = '''一生一世
二三三
不三不四'''

re.findall(r'[零〇一二三四五六七八九十廿卅卌百皕千]', text)

# правда я не знаю, насколько это исчерпывающий список чисел...
```

```python
# где-то прочитала, что извлечь вообще все китайские иероглифы разом можно например так:

# китайские фразы из цифр
text = '''1314 yī sān yī sì
一生一世 yīshēng yīshì
На всю жизнь.

360 sān liù líng
想念你 xiǎngniàn nǐ
Скучаю по тебе.

25184 èr wǔ yī bā sì
爱我一辈子 ài wǒ yībèizi
Люби меня всю жизнь.'''

re.findall(r'[\u4e00-\u9fff]+', text)
# где \u4e00 - первый иероглиф в диапазоне
# \u9fff - последний
```
