- Через пробел вводятся пароли.

# Задача 8

Реализуйте программу, которая считает количество цен на картины, которые больше определенного значения:
- На первой строке через пробел вводятся цены.
- На второй строке вводится интересующее нас значение (целое число) в переменной price.
- Программа должна учитывать только те цены, которые больше введенного значения price. Для этого нужно перебрать переменную pictures циклом for, внутри цикла for создать проверку условия - if, и проверять, что очередная цена больше price. В этой задаче не печатайте ее, а сохраните в переменную count, увеличив ее на 1 (этот код уже написан).
- В конце программа печатает количество цен, которые больше введенного значения.

```python
pictures = '18293 18000 15462 19222'.split()
price = int(input()) # введите сюда любую цену для проверки работы программы
count = 0 # сколько картин, чья цена выше price

# ваше решение здесь
# for
# if
    count += 1 # это значит: увеличь счетчик картин, которые стоят выше price, на 1
print(count)
```

```python
# @title
for i in pictures:
  if int(i) > price:
    count += 1 # увеличиваем счетчик дорогих картин
print(count) # печатаем после цикла, сколько таких картин мы нашли
```
